## Data Visualization


## 1. Based on category, segment, ship mode and region


```python
fig,axs=plt.subplots(nrows=2,ncols=2,figsize=(10,7));

sns.countplot(df['Category'],ax=axs[0][0])

sns.countplot(df['Segment'],ax=axs[0][1])

sns.countplot(df['Ship Mode'],ax=axs[1][0])

sns.countplot(df['Region'],ax=axs[1][1])

axs[0][0].set_title('Category',fontsize=20)

axs[0][1].set_title('Segment',fontsize=20)

axs[1][0].set_title('Ship Mode',fontsize=20)

axs[1][1].set_title('Region',fontsize=20)

plt.tight_layout()
```

## 2. Based on sub category 

```python
plt.figure(figsize=(20,8))

sns.countplot(df['Sub-Category'])

plt.title('Sub-Category',fontsize=20)
```

## Based on Quantity

```python
plt.figure(figsize=(20,8))

sns.countplot(df['Quantity'])

plt.title('Quantity',fontsize=20)
```

## Based on Discount

```python
plt.figure(figsize=(20,8))

sns.countplot(df['Discount'])

plt.xticks(rotation=90)

plt.title('Discount',fontsize=20)
```

## Find correlation of data

```python
sns.heatmap(df.corr(),annot=True)
```

From the above heatmap, we can say that sales and profit are moderately correlated. Discount and profit are negatively correlated. Quantity and profit are less moderately correlated

## Find distribution of data 

```python
fig, axs = plt.subplots(ncols=2, nrows = 2, figsize = (10,10))

sns.distplot(df['Sales'], color = 'red',  ax = axs[0][0])

sns.distplot(df['Profit'], color = 'green',  ax = axs[0][1])

sns.distplot(df['Quantity'], color = 'orange',  ax = axs[1][0])

sns.distplot(df['Discount'], color = 'blue',  ax = axs[1][1])

axs[0][0].set_title('Sales Distribution', fontsize = 20)

axs[0][1].set_title('Profit Distribution', fontsize = 20)

axs[1][0].set_title('Quantity distribution', fontsize = 20)

axs[1][1].set_title('Discount Distribution', fontsize = 20)

plt.show()
```

Here, sales distribution is rightly skewed and profit distribution is normally skewed

## State wise analysis

## 1. based on dealings

```python
df1 = df['State'].value_counts()

df1.plot(kind='bar',figsize=(15,5), color='orange')

plt.ylabel('Frequency / Number of deals')

plt.xlabel('States')

plt.title('Dealings', fontsize = 20)

plt.show()
```

From the bar graph, we can say that California has the highest number of deals. States such as Maine, North Dakota and West Virginia has little to no less number of deals.

## 2. based on  profit

```python
df_state= df.groupby(['State'])[['Sales', 'Discount', 'Profit']].mean()

df_state1=df_state.sort_values('Profit')

df_state1[['Profit']].plot(kind = 'bar', figsize = (15,4), color='orange')

plt.title('Profit Analysis', fontsize = 20)

plt.ylabel('Profit per State')

plt.xlabel('States')

plt.show()
```

From the bar graph, we can say that Vermont has the highest profit and Ohio has the lowest profit

## 3.based on sale 

```python
df_state1[['Sales']].plot(kind = 'bar', figsize = (15,4), color='orange')

plt.title('Sale Analysis', fontsize = 20)

plt.ylabel('Profit per State')

plt.xlabel('States')

plt.show()
```

The highest amount of sales is observed in Wyoming and the lowest amount of sales is seen in South Dakota.

## 4. based on discount 

```python
df_state1['Discount'].plot(kind='bar',figsize=(18,5), color='orange')

plt.title('Discount analysis', fontsize=20)
```

Illinois has the highest discount

## City wise dealings

## 1. based on dealings

```python
df2 = df['City'].value_counts()

df2=df2.head(50)

df2.plot(kind='bar',figsize=(15,5), color='pink')

plt.ylabel('Frequency / Number of deals')

plt.xlabel('City')

plt.title('Dealings', fontsize = 20)

plt.show()
```

The highest number of deals is found in New York City

## 2. based on profit 

```python
df_city= df.groupby(['City'])[['Sales', 'Discount', 'Profit']].mean()

df_city1=df_city.sort_values('Profit')

df_city1[['Profit']].head(50).plot(kind = 'bar', figsize = (15,4), color='pink')

plt.title('Low profit', fontsize = 20)

plt.ylabel('Profit per city')

plt.xlabel('city')

plt.show()
```

```python
df_city= df.groupby(['City'])[['Sales', 'Discount', 'Profit']].mean()

df_city1=df_city.sort_values('Profit')

df_city1[['Profit']].tail(50).plot(kind = 'bar', figsize = (15,4), color='pink')

plt.title('High profit', fontsize = 20)

plt.ylabel('Profit per city')

plt.xlabel('city')

plt.show()
```

We can conclude that 50 cities have profit in positive and 50 cities have profit in negative and hence balanced

## 3. based on sale 

```python
df_city1[['Sales']].head(50).plot(kind = 'bar', figsize = (15,4), color='pink')

plt.title('Sale Analysis', fontsize = 20)

plt.ylabel('sales per city')

plt.xlabel('city')

plt.show()
```

Burlington has the highest number of sales and Skokie has the lowest number of sales

## 4. based on discount 

```python
df_city1['Discount'].head(50).plot(kind='bar',figsize=(18,5), color = 'pink')

plt.title('Discount Analysis', fontsize=20)
```

Littleton has the highest discount

## Quantity wise analysis

```python
df_quantity = df.groupby(['Quantity'])[['Sales', 'Discount', 'Profit']].mean()

df_quantity.plot.pie(subplots=True, autopct='%1.1f%%', figsize=(20, 20), pctdistance=0.69,
                     startangle=90, shadow=True, labels = df_quantity.index)

plt.title('Quantity wise analysis of Sale, Discount, Profit')
```

13 number of quantity is high for sales and profit

## Category wise analysis

```python
df_category = df.groupby(['Category'])[['Sales', 'Discount', 'Profit']].mean()

df_category.plot.pie(subplots=True, figsize=(18, 20), autopct='%1.1f%%', labels = df_category.index)
```

Maximun sales and profit is obtained in Technology
Minimun profit is obtained in Furniture

## Region wise analysis

```python
df_region = df.groupby(['Region'])[['Sales', 'Discount', 'Profit']].mean()

df_region.plot.pie(subplots=True, figsize=(18, 20), autopct='%1.1f%%', labels = df_region.index)
```

Profit is high in the west region

## ship mode wise analysis

```python
df_shipmode = df.groupby(['Ship Mode'])[['Sales', 'Discount', 'Profit']].mean()

df_shipmode.plot.pie(subplots=True, figsize=(18, 20), autopct='%1.1f%%', labels = df_shipmode.index)
```

Profit and discount is high in First Class.
Sales is high for Same day ship

## Interdependency of sales, profits and discounts

```python
df.plot(kind = "scatter",figsize = (10,5), x = "Sales", y= "Profit", c = "Discount", s = 20,fontsize = 16, colormap = "viridis")

plt.ylabel('Total Profits',fontsize = 16)

plt.title("Interdependency of Sales,Profits and Discounts",fontsize = 16)

plt.show()
```

```python
plt.figure(figsize=(8,7))

sns.lineplot(df['Discount'], df['Profit'], data=df)
```

The percentage of discount should be less than 25% to maintain the profit margins
